# Call Binance public Api and create a csv dataset using python and pandas

```python
import binance
import pandas as pd
```

```python
#getting data from api
client=binance.Client()
tick=client.get_ticker()
tick
```

```python
#converting the ticker data to pandas dataframe
tick_df=pd.DataFrame(tick)
tick_df
```

```python
#checking features datatypes
tick_df.dtypes
```

```python
#converting features data type from object to int or float for ease of calculations
feat=[tick_df.columns]
feature_conv=['priceChange', 'priceChangePercent', 'weightedAvgPrice',
        'prevClosePrice', 'lastPrice', 'lastQty', 'bidPrice', 'bidQty',
        'askPrice', 'askQty', 'openPrice', 'highPrice', 'lowPrice', 'volume',
        'quoteVolume', 'openTime', 'closeTime', 'firstId', 'lastId', 'count']
for i in feature_conv:
    tick_df[i]=pd.to_numeric(tick_df[i])
tick_df.dtypes
```

```python
#saving the dataframe to a csv file
tick_df.to_csv("D:/binance_data.csv")
```
