# To create plots using Matplotlib.pyplot and seaborn plots for the given dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv("D:/CodesOnBytes/dataset - netflix1.csv")
```

```python
df
```

```python
#getting year and no of movies/TV Shows released in that year
year=list()
for i in df.release_year:
    if i not in year:
        year.append(i)
year_count=[]
for j in year:
    count=0
    for k in df.release_year:
        if j==k:
            count+=1
    year_count.append(count)
```

```python
#pie chart of the release years of movies/TV Shows present in netflix 
palette_color=sns.color_palette('bright')
plt.figure(figsize=(30,20))
plt.pie(year_count,labels=year,colors=palette_color, autopct='%.0f%%')
plt.show()
```

```python
#scatter plot of No of movies/TV Shows vs its release year
import matplotlib.pyplot as plt1
plt1.figure(figsize=(10,5))
plt1.scatter(year,year_count)
plt1.xlabel('Release Year')
plt1.ylabel('No of Movies/TV Shows')
plt1.title('Movies/TV Shows vs Release year')
plt1.show()
```

```python
#No of Movies in Netflix
type_count=list()
type_count.append(len(df[df['type']=='Movie']))
type_count.append(len(df[df['type']=='TV Show']))
```

```python
#bar graph for type of content in netflix
content=['Movie','TV Show']
plt.figure(figsize=(6,3))
con=sns.barplot(x=content,y=type_count,width=0.2)
con.set(xlabel='Type',ylabel='count',title='Type of Content and its Count')
plt.show()
```

```python
#To find the no of movies/ TV Show for each kind of rating.
rating=list()
for i in df.rating:
    if i not in rating:
        rating.append(i)
```

```python
rating_count=list()
for i in rating:
    count=0
    for j in df.rating:
        if i==j:
            count+=1
    rating_count.append(count)
```

```python
import matplotlib.pyplot as plt2
plt2.figure(figsize=(15,10))
rate=sns.barplot(x=rating,y=rating_count,width=0.5)
rate.set(xlabel='ratings',ylabel='No of Movies', title='No of content vs Type of Ratings')
plt2.show()
```

```python
df_m=df[df['type']=='Movie']
```

```python
# creating a dictionary with {director:no of movies} key-value pair
director_m=dict()
for i in df_m.director:
    if i not in director_m:
        director_m[i]=1
    else:
        director_m[i]+=1
```

```python
#creating a new sorted dictionary with descending order of film number
sort_director_m=dict(sorted(director_m.items(), key=lambda item: item[1], reverse=True))
```

```python
mdir_name=list(sort_director_m)[:4]
mdir_count=list(sort_director_m.values())[:4]
```

```python
mdir_name=mdir_name[1:]
mdir_count=mdir_count[1:]
```

```python
mdir_count
```

```python
import matplotlib.pyplot as plt3
plt3.figure(figsize=(8,4))
mdir=sns.barplot(x=mdir_name,y=mdir_count,width=0.5)
mdir.set(xlabel='Name',ylabel='No of Movies',title='Top 4 directors with High Movie count in Netflix')
plt3.show()
```
